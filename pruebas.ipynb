```python

from fastapi import FastAPI
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import fastparquet
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import GradientBoostingRegressor
import joblib

# Nombre del archivo Joblib
model_filename = 'gradient_boosting_model.joblib'

# Cargar el modelo, StandardScaler y LabelEncoder
with open(model_filename, 'rb') as file:
    model, scaler, encoder = joblib.load(file)

# Se carga el dataframe
df = pd.read_parquet('steam_games.parquet')

app = FastAPI()



import matplotlib.pyplot as plt

# Función para obtener los géneros más vendidos
@app.get("/genero/")
def genero(anio: int):
    df_filtrado = df[df['release_date'] == anio]
    if df_filtrado.empty:
        return {"!": f"No hay datos disponibles para el año {anio}"}
    
    generos_count = {}
    for lista_generos in df_filtrado['genres']:
        if lista_generos is not None:
            for genero_juego in lista_generos:
                if genero_juego is not None:
                    generos_count[genero_juego] = generos_count.get(genero_juego, 0) + 1
    
    generos_top = dict(sorted(generos_count.items(), key=lambda item: item[1], reverse=True)[:5])

    # Generar el gráfico de barras
    etiquetas = generos_top.keys()
    valores = generos_top.values()

    plt.figure(figsize=(10, 6))  # Tamaño del gráfico
    plt.barh(etiquetas, valores, color='skyblue')  # Gráfico de barras horizontales
    plt.xlabel("Cantidad")
    plt.ylabel("Géneros")
    plt.title(f"Géneros más vendidos en el año {anio}")
    
    # Agregar etiquetas con los valores en las barras
    for i, v in enumerate(valores):
        plt.text(v + 1, i, str(v), ha='left', va='center', fontsize=10, color='black')
    
    plt.tight_layout()  # Ajustar el diseño
    plt.show()

    return generos_top
```
